```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
cb_grey = (171/255, 171/255, 171/255)

stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering', 'Computer Science']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
degree_cats = [stem_cats, lib_arts_cats, other_cats]

fig = plt.figure(figsize=(16, 16))

for i in range(0, 3):
    x = 15 if i == 1 else 18
    for sp in range(i, x, 3):
        ax = fig.add_subplot(6, 3, sp+1)
        ax.plot(women_degrees['Year'], women_degrees[degree_cats[i][int((sp-i)/3)]], label='Women', c=cb_dark_blue, linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[degree_cats[i][int((sp-i)/3)]], label='Men', c=cb_orange, linewidth=3)
        ax.set_xlim(1968, 2011)
        ax.set_ylim(0, 100)
        ax.set_title(degree_cats[i][int((sp-i)/3)])
        ax.tick_params(top='off', bottom='off', left='off', right='off', labelbottom='off')
        for key, spine in ax.spines.items():
            spine.set_visible(False)
        anno_dict = {1: [[2003,90], [2003,30]],
                     2: [[2003,90], [2003,40]],
                     3: [[2003,90], [2003,30]],
                    16: [[2003,30], [2003,90]],
                    18: [[2003,10],  [2003,80]]}
        for key, coord in anno_dict.items():
            if sp + 1 == key:
                ax.text(coord[0][0], coord[0][1], 'Women')
                ax.text(coord[1][0], coord[1][1], 'Men')
        if sp == 13 or sp == 15 or sp == 17:
            ax.tick_params(labelbottom='on')
        ax.set_yticks([0,100])
        ax.axhline(50, c=cb_grey, alpha=0.3)
plt.savefig('gender_degrees.png')         
plt.show()
    
```
